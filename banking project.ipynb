```python
import sqlite3
```

```python
try:
    con=sqlite3.connect(database="E:\ banking.sqlite")
    cur=con.cursor()
    cur.execute("create table accounts(acn integer primary key autoincrement,name text,password text,email text,mob text,bal float,type text,opendate text)")
    cur.execute("create table txn_history(acn int,txn_amt float,txn_type float,txn_date text,update_bal float)")
    con.commit()
    print("tables created")
except:
    print("something went wrong,migh be tables already exist")
    con.close()
```

```python
from tkinter import *
from datetime import datetime
from tkinter.ttk import Combobox
```

```python
win=Tk()
win.state("zoomed")
win.resizable(width=False,height=True)
win.configure(bg='skyblue')
label_title=Label(win,text="State Bank of India",bg='skyblue',fg='white',font=('green',60,'underline'))
label_title.pack()

lbl_date=Label(win,text=f"{datetime.now().date()}",bg='powder blue',font=('arial',15,'bold'))
lbl_date.place(relx=.9,rely=.1)

def main_screen():
    frm=Frame(win)
    frm.configure(bg='pink')
    frm.place(x=0,rely=.16,relwidth=1,relheight=.85)

    def login_account():
        frm.destroy()
        loginaccount_screen()

    def forgot_pass():
        frm.destroy()
        forgotpass_screen()

    def newacn():
        frm.destroy()
        newacn_screen()

    lbl_acn=Label(frm,text="Account No.",bg='pink',font=('arial',20,'bold'))
    lbl_acn.place(relx=.3,rely=.1)

    entry_acn=Entry(frm,font=('arial',20,'bold'),bd=5)
    entry_acn.place(relx=.45,rely=.1)

    lbl_pass=Label(frm,text="Password:",bg='pink',font=('arial',20,'bold'))
    lbl_pass.place(relx=.3,rely=.21)
  
    entry_pass=Entry(frm,font=('arial',20,'bold'),bd=5)
    entry_pass.place(relx=.45,rely=.2)

    login_btn=Button(frm,command=login_account,text="Login",bg='pink',font=('arial',20,'bold'))
    login_btn.place(relx=.45,rely=.3)

    pass_btn=Button(frm,command=forgot_pass,text="Forgot Password",bg='pink',font=('arial',20,'bold'))
    pass_btn.place(relx=.45,rely=.4)

    newacn_btn=Button(frm,command=newacn,text="Create New Account",bg='pink',font=('arial',20,'bold'))
    newacn_btn.place(relx=.45,rely=.5)

def loginaccount_screen():
    frm=Frame(win)
    frm.configure(bg='pink')
    frm.place(x=0,rely=.15,relwidth=1,relheight=.85) 

    def logout():
        frm.destroy()
        main_screen()

    def details():
        inner_frm=Frame(frm,highlightthickness=2,highlightbackground='black')
        inner_frm.configure(bg='white')
        inner_frm.place(relx=.25,rely=.1,relwidth=.55,relheight=.5)
        lbl_frmtitle.configure(text="This is Check Details Screen")
        lbl_frmtitle.pack()

    def updateprofile():
        inner_frm=Frame(frm,highlightthickness=2,highlightbackground='black')
        inner_frm.configure(bg='white')
        inner_frm.place(relx=.25,rely=.1,relwidth=.55,relheight=.5)
        lbl_frmtitle.configure(text="This is Update Details Screen")
        lbl_frmtitle.pack()

    def deposit():
        inner_frm=Frame(frm,highlightthickness=2,highlightbackground='black')
        inner_frm.configure(bg='white')
        inner_frm.place(relx=.25,rely=.1,relwidth=.55,relheight=.5)
        lbl_frmtitle.configure(text="Earn more Deposit More")
        lbl_frmtitle.pack()

    def withdraw():
        inner_frm=Frame(frm,highlightthickness=2,highlightbackground='black')
        inner_frm.configure(bg='white')
        inner_frm.place(relx=.25,rely=.1,relwidth=.55,relheight=.5)
        lbl_frmtitle.configure(text="Withdraw less Invest More")
        lbl_frmtitle.pack()

    def txnhistory():
        inner_frm=Frame(frm,highlightthickness=2,highlightbackground='black')
        inner_frm.configure(bg='white')
        inner_frm.place(relx=.25,rely=.1,relwidth=.55,relheight=.5)
        lbl_frmtitle.configure(text="Your Spendings Are:")


    
    btn_back=Button(frm,command=logout,text="logout",font=('arial',15,'bold'),bd=5,bg='powderblue')
    btn_back.place(relx=.9,rely=0)
    
    lbl_frmtitle=Label(frm,text="This is Login Account Screen",bg='pink',font=('arial',20,'bold'),fg='green')
    lbl_frmtitle.place(relx=.3,rely=.05)
    
    lbl_wel=Label(frm,text="Welcome, Prince",bg='pink',font=('arial',20,'bold'),fg='green')
    lbl_wel.place(relx=0,rely=0)

    btn_details=Button(frm,command=details,width=15,text="Check details",font=('arial',18,'bold'),bd=5,bg='powderblue')
    btn_details.place(relx=0,rely=.3)

    btn_updateprofile=Button(frm,command=updateprofile,width=15,text="Update Profile",font=('arial',18,'bold'),bd=5,bg='powderblue')
    btn_updateprofile.place(relx=0,rely=.4)

    btn_deposit=Button(frm,command=deposit,width=15,text="Deposit",font=('arial',18,'bold'),bd=5,bg='powderblue')
    btn_deposit.place(relx=0,rely=.5)

    btn_withdraw=Button(frm,command=withdraw,width=15,text="Withdraw",font=('arial',18,'bold'),bd=5,bg='powderblue')
    btn_withdraw.place(relx=0,rely=.6)

    btn_txnhistory=Button(frm,width=15,command=txnhistory,text="Transaction History",font=('arial',18,'bold'),bd=5,bg='powderblue')
    btn_txnhistory.place(relx=0,rely=.7)

def forgotpass_screen():
    frm=Frame(win)
    frm.configure(bg='pink')
    frm.place(x=0,rely=.15,relwidth=1,relheight=.85)

    def back():
        frm.destroy()
        main_screen()

    lbl_frmtitle=Label(frm,text="Forgot Password Screen",bg='pink',fg='green',font=('arial',20,'bold'))
    lbl_frmtitle.pack()

    lbl_acn=Label(frm,text="ACN",bg='pink',font=('arial',17,'bold'))
    lbl_acn.place(relx=.3,rely=.1)
    
    entry_acn=Entry(frm,font=('arial',17,'bold'),bd=4)
    entry_acn.place(relx=.55,rely=.1)
    entry_acn.focus()

    lbl_recemail=Label(frm,text="Enter Your Email Id:",bg='pink',font=('arial',17,'bold'))
    lbl_recemail.place(relx=.3,rely=.2)

    entry_email=Entry(frm,font=('arial',17,'bold'),bd=4)
    entry_email.place(relx=.55,rely=.2)

    btn_otp=Button(frm,text="Send OTP",font=('arial',15,'bold'),bd=4,bg='powderblue')
    btn_otp.place(relx=.55,rely=.3)

    btn_back=Button(frm,command=back,text="Return to Home Page",font=('arial',15,'bold'),bd=4,bg='powderblue')
    btn_back.place(relx=.82,rely=0)

def newacn_screen():
    frm=Frame(win)
    frm.configure(bg='pink')
    frm.place(x=0,rely=.15,relwidth=1,relheight=.85)

    def back():
        frm.destroy()
        main_screen()

    lbl_1stname=Label(frm,text="First Name:",bg='pink',font=('arial',18,'bold'))
    lbl_1stname.place(relx=.3,rely=.1)

    entry_1stname=Entry(frm,font=('arial',18,'bold'),bd=5)
    entry_1stname.place(relx=.5,rely=.1)

    lbl_2ndname=Label(frm,text="Last Name:",bg='pink',font=('arial',18,'bold'))
    lbl_2ndname.place(relx=.3,rely=.2)

    entry_2ndname=Entry(frm,font=('arial',18,'bold'),bd=5)
    entry_2ndname.place(relx=.5,rely=.2)

    lbl_email=Label(frm,text="Email id:",bg='pink',font=('arial',18,'bold'))
    lbl_email.place(relx=.3,rely=.3)

    entry_email=Entry(frm,font=('arial',18,'bold'),bd=5)
    entry_email.place(relx=.5,rely=.3)

    lbl_mob=Label(frm,text="Mob No.",bg='pink',font=('arial',18,'bold'))
    lbl_mob.place(relx=.3,rely=.4)

    entry_mob=Entry(frm,font=('arial',18,'bold'),bd=5)
    entry_mob.place(relx=.5,rely=.4)

    lbl_pswrd=Label(frm,text="Enter new Password:",bg='pink',font=('arial',18,'bold'))
    lbl_pswrd.place(relx=.3,rely=.5)

    entry_pswrd=Entry(frm,font=('arial',18,'bold'),bd=5)
    entry_pswrd.place(relx=.5,rely=.5)

    lbl_acnttype=Label(frm,text="Choose Account Type:",bg='pink',font=('arial',18,'bold'))
    lbl_acnttype.place(relx=.3,rely=.6)

    combo_type=Combobox(frm,font=('arial',18,'bold'),values=['Saving','Current'])
    combo_type.current(0)
    combo_type.place(relx=.5,rely=.6)

    btn_back=Button(frm,command=back,text="Back",font=('arial',12,'bold'),bd=5,bg='powderblue')
    btn_back.place(relx=0,rely=0)



main_screen()
win.mainloop()
```
