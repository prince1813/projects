```python
from tkinter import *
```

```python
win=Tk()
win.state("zoomed")
win.resizable(width=False,height=True)
win.configure(bg='pink')

lbl_title=Label(win,
                text="State bank of India",
                bg='pink',fg='black',
                font=('arial',60,'bold','underline'))
lbl_title.pack()

def main_screen():
    frm=Frame(win)
    frm.configure(bg='light blue')
    frm.place(x=0,rely=.15,relwidth=1,relheight=0.85)

    def forgot_pass():
        frm.destroy()
        forgotpass_screen()

    def open_account():
        frm.destroy()
        openaccount_screen()

    def login_account():
        frm.destroy()
        loginaccount_screen()

    lbl_acn=Label(frm,text="Acc.No.",bg='light blue',font=('arial',20,'bold'))
    lbl_acn.place(relx=.3,rely=.1)

    entry_acn=Entry(frm,font=('arial',20,'bold'),bd=5)
    entry_acn.place(relx=.4,rely=.1)
    entry_acn.focus()

    lbl_pswrd=Label(frm,text="Password",bg='light blue',font=('arial',20,'bold'))
    lbl_pswrd.place(relx=.3,rely=.2)

    entry_pswrd=Entry(frm,font=('arial',20,'bold'),bd=5,show="*")
    entry_pswrd.place(relx=.4,rely=.2)

    btn_login=Button(frm,command=login_account,
                     text="Login",
                     bg='pink',font=('arial',20,'bold'),bd=2,width=8)
    btn_login.place(relx=.4,rely=.3)

    btn_reset=Button(frm,text="Reset",bg='pink',font=('arial',20,'bold'),bd=2,width=8)
    btn_reset.place(relx=.51,rely=.3)

    btn_forgotpswrd=Button(frm,command=forgot_pass,
                           text="Forgot Password",
                           bg='pink',font=('arial',20,'bold'),bd=2,width=18)
    btn_forgotpswrd.place(relx=.4,rely=.4)

    btn_createacnt=Button(frm,command=open_account,
                          text="Create New Account",bg='pink',font=('arial',20,'bold'),bd=2,width=18)
    btn_createacnt.place(relx=.4,rely=.5)

def forgotpass_screen():
    frm=Frame(win)
    frm.configure(bg='light blue')
    frm.place(x=0,rely=.15,relwidth=1,relheight=0.85)

    def back():
        frm.destroy()
        main_screen()

    btn_back=Button(frm,command=back,text="Back",bg='pink',font=('arial',20,'bold'),bd=2,width=8)
    btn_back.place(relx=0,rely=.01)

    lbl_frmtitle=Label(frm,text="Forgot Password Screen",bg='light blue',font=('arial',20,'bold'),fg='green')
    lbl_frmtitle.place(relx=.3,rely=.02)


def openaccount_screen():
    frm=Frame(win)
    frm.configure(bg='light blue')
    frm.place(x=0,rely=.15,relwidth=1,relheight=0.85)

    def back():
        frm.destroy()
        main_screen()

    btn_back=Button(frm,command=back,text="Back",bg='pink',font=('arial',20,'bold'),bd=2,width=8)
    btn_back.place(relx=0,rely=.01)

    lbl_frmtitle=Label(frm,text="Open Account Screen",bg='light blue',font=('arial',20,'bold'),fg='green')
    lbl_frmtitle.place(relx=.3,rely=.02)

def loginaccount_screen():
    frm=Frame(win)
    frm.configure(bg='light blue')
    frm.place(x=0,rely=.15,relwidth=1,relheight=0.85)

    def back():
        frm.destroy()
        main_screen()

    btn_back=Button(frm,command=back,text="Back",bg='pink',font=('arial',20,'bold'),bd=2,width=8)
    btn_back.place(relx=0,rely=.01)

    lbl_frmtitle=Label(frm,text="Login Account Screen",bg='light blue',font=('arial',20,'bold'),fg='green')
    lbl_frmtitle.place(relx=.3,rely=.02)



main_screen()
win.mainloop()
```
